```python
#importing the libraries 
import pandas as pd
import matplotlib.pyplot as plt
#Loading the data set
df=pd.read_csv("C:\\Users\\DELL\\Desktop\\data science\\netflix data analysis\\netflix_titles.csv",encoding="latin1")

#cleaning dataset
df= df.dropna(subset=['type','release_year','rating','country','duration'])
#counting types in data set
Type_count=df["type"].value_counts()
plt.Figure(figsize=(3,2))
plt.bar(Type_count.index,Type_count.values,color=["red","blue"])
plt.title("Number of Movies  vs Tv show on Netflix")
plt.xlabel("type of content")
plt.ylabel("content avaible")
plt.tight_layout()
plt.savefig("movies vs tv shows on netflix.png")
plt.show()
```

```python
#counting ratings percentage
rating_count = df["rating"].value_counts()
plt.Figure(figsize=(10,8))
plt.pie(rating_count, labels=rating_count.index,autopct="%1.1f%%",startangle=90)
plt.title("percentage of content rating")
plt.tight_layout()
plt.savefig("Percentage of rating.png")
plt.show()
```

```python
#counting  number of movies in each ratings  using bar chart
rating_count = df["rating"].value_counts()
plt.Figure(figsize=(10,8))
plt.bar(rating_count.index,rating_count,color="orange")
plt.title("percentage of content rating")
plt.tight_layout()
plt.savefig("Percentage of rating.png")
plt.show()
```

```python
release_year_count=df["release_year"].value_counts()
plt.figure(figsize=(10,6))
plt.scatter(release_year_count.index,release_year_count,color="red",marker="o")
plt.title("Number of Movies relases in each year")
plt.xlabel("Years")
plt.ylabel("Number of movies")
plt.tight_layout()
plt.savefig("number of movies releases in each year")
plt.show()
```

```python
country_counts=df["country"].value_counts().head(10)
plt.Figure(figsize=(8,6))
plt.barh(country_counts.index,country_counts)
plt.title("Top 10 Countries by number of shows")
plt.xlabel("Number of Show releases")
plt.ylabel("Countries")
plt.tight_layout()
plt.savefig("Top 10 countries by number of shows")
plt.show()
```

```python
content_by_year=df.groupby(["release_year", "type"]).size().unstack().fillna(0)
fig,ax=plt.subplots(1,2,figsize=(10,6))
#first subplot::Movies
ax[0].plot(content_by_year,content_by_year["Movie"],color="blue")
ax[0].set_title("movies released per year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("number of movies ")

#second subplot::Tv show
ax[1].plot(content_by_year,content_by_year["TV Show"],color="Orange")
ax[1].set_title("Tv shows released per year")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("number of Tv shows")

fig.suptitle("comparison of movies and Tv show  release over year")
plt.tight_layout()
plt.savefig("Movies_tv_show comparison.png")

plt.show()
```
